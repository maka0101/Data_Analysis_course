Для $U[0, 1]$ сгенерируйте $200$ независимых случайных величин, постройте график плотности на отрезке $[-0.25, 1.25]$, а также гистограмму по сгенерированным случайным величинам. Для построения графика необходимо брать достаточно большое количество точек сетки.

```python
from scipy.stats import uniform
uniform_ = uniform.rvs(size=200, loc=0, scale=1)

x = np.linspace(-0.25, 1.25, 1000)
theory = uniform.pdf(x, loc=0, scale=1)

plt.figure(figsize=(12, 6))

plt.hist(uniform_,
         bins=20,
         density=True,
         alpha=0.5,
         edgecolor='black',
         label='Гистограмма')

plt.plot(x, theory,
        'r-',
        linewidth=2,
        label='Теоретическая плотность')

plt.title('Равномерное распределение U[0,1]', fontsize=14)
plt.xlabel('Значение', fontsize=12)
plt.ylabel('Плотность', fontsize=12)
plt.xlim(-0.25, 1.25)
plt.ylim(0, 1.5)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)

plt.show()
```

Исследуйте, как меняются значения случайных величин в зависимости от `precision` в написанной ранее вами функции.

```python
from scipy.stats import uniform
import scipy.stats as sps

coin = sps.bernoulli.rvs(p=0.5)

def coin(n):
    return sps.bernoulli.rvs(p=0.5, size=n)

def uniform_(size=1, precision=30):
  size_of_res = np.prod(size)

  binary = coin(precision * size_of_res).reshape(size_of_res, precision)
  pows_of_2 = 0.5 ** (np.arange(1, precision + 1))

  return np.dot(binary, pows_of_2).reshape(size)

def plot_uniform_precision_impact(sample_size=10000, precisions=(5, 10, 20, 30, 40, 50)):
    plt.figure(figsize=(15, 15))

    x = np.linspace(-0.25, 1.25, 1000)
    theory = uniform.pdf(x, loc=0, scale=1)

    for i, precision in enumerate(precisions, 1):
        data = uniform_(size=sample_size, precision=precision)

        # Построение гистограммы
        plt.subplot(3, 2, i)
        plt.hist(data, bins=50, density=True, alpha=0.6,
                label=f'precision={precision}')
        plt.plot(x, theory, 'r-', lw=2, label='Теоретическая плотность')

        # Оформление
        plt.title(f'precision = {precision}', fontsize=12)
        plt.xlabel('Значение', fontsize=10)
        plt.ylabel('Плотность', fontsize=10)
        plt.xlim(-0.1, 1.1)
        plt.ylim(0, 1.5)
        plt.grid(alpha=0.3)
        plt.legend()

    plt.tight_layout()
    plt.show()

# Запуск визуализации
plot_uniform_precision_impact()
```

За распределение Коши отвечает класс <a target="_black" href="https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.cauchy.html">`scipy.stats.cauchy`</a>.

**2.** Повторите эксперимент с <a href="https://thetahat.ru/courses/ad/main/9/lec9_LLN" target="_top">занятия</a> про закон больших чисел для распределения Коши. Выполняется ли закон?

```python
from scipy.stats import cauchy
size = 1000
samples_count = 500

distr = sps.cauchy()
ylim = (-10, 10)
label = "$Cauchy(0, 1)$"

samples = distr.rvs(size=(samples_count, size))
cum_means = samples.cumsum(axis=1) / (np.arange(size) + 1)

# График
plt.figure(figsize=(15, 7))
for i in range(samples_count):
    plt.plot(np.arange(size) + 1, cum_means[i], color="green", alpha=0.05)
plt.title("$\\xi_i \\sim " + label[1:], fontsize=16)
plt.xlabel("Количество случайных величин")
plt.ylabel("Значение среднего")
plt.xlim((0, size))
plt.ylim(ylim)
plt.show()
```

Как видно, из-за этих "волосков" на графике, на распределении Коши ЗБЧ ломается(в частности из-за отсутствия матожа, но на графике это показано наглядно).

**3.** Аналогичным образом проведите эксперимент по изучению зависимости *выборочной* медианы от размера выборки. Медиану по выборке можно посчитать с помощью <a target="_blank" href="https://numpy.org/doc/stable/reference/generated/numpy.median.html">`np.median`</a>.

```python
from scipy.stats import cauchy

sample = np.arange(10, 1001, 10)
rep = 300

cauchy_medians = np.zeros(len(sample), dtype=object)
cauchy_mean = [np.mean(m) for m in cauchy_medians]

for i, size in enumerate(sample):
    medians = []
    for _ in range(rep):
        sample_ = cauchy.rvs(size=size)
        medians.append(np.median(sample_))

    cauchy_medians[i] = medians
    cauchy_mean[i] = np.mean(medians)


# Визуализация
plt.figure(figsize=(12, 6))

# График средних медиан
plt.plot(
    sample,
    cauchy_mean,
    color='#0055ff',
    lw=3,
    marker='o',
    markersize=4,
    markevery=20,
    label='Средняя медиана',
    zorder=3
)

# Теоретическая медиана
plt.axhline(
    0,
    color='#ff3300',
    linestyle='--',
    lw=2.5,
    label='Теоретическая медиана (0)',
    zorder=2
)

# Настройки масштаба и оформления
plt.ylim(-0.3, 0.3)
plt.xlim(0, 1000)
plt.yticks(np.arange(-0.3, 0.31, 0.1), fontsize=12)

plt.title('Зависимость медианы от размера выборки',
         fontsize=14, pad=15)
plt.xlabel('Размер выборки', fontsize=12)
plt.ylabel('Медиана', fontsize=12)

plt.grid(alpha=0.2, linestyle='--')
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```

**Вывод:**

Так, в этой задаче изучили распределение Коши. Проверили выполнимость ЗБЧ для него(как видно из графика, не работает ЗБЧ на распределнии Коши).

+ Наглядно показали тяжёлые хвосты распределения Коши по сравнению с нормальным распределением.

+ При увеличении выборки, распределение почти совпадает с медианным значением.
